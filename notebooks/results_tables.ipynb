```python
import pandas as pd


df = pd.read_csv('wandb_export_4ds.csv')
df
```

```python
df_part = df[['Name', 'best_test_acc', 'dataset']]
# df_part['dataset'] = 'K-MNIST'
df_part['directions'] = df_part['Name'].str.split().str[1]
df_part['num_directions'] = df_part['directions'].str.count(',') + 1
df_part = df_part[['dataset', 'num_directions', 'directions', 'best_test_acc']]
df_part.sort_values(by='best_test_acc', ascending=False).reset_index(drop=True)

# df_part
```

```python
df_grouped = df_part.groupby(['dataset', 'num_directions'])['best_test_acc'].max().reset_index().sort_values(by='best_test_acc', ascending=False)
df_pivoted = df_grouped.pivot(index='dataset', columns='num_directions', values='best_test_acc')
df_pivoted = df_pivoted[[1, 2, 4, 8]]
df_pivoted
```

```python
df_grouped = df_part.groupby('num_directions')['best_test_acc'].max().reset_index().sort_values(by='best_test_acc', ascending=False)
df_grouped
```

```python
import wandb
runs_to_update = {
    "sfw983o1": 'MNIST', 
    "k8hxujgk": 'MNIST',
    '88bglsn9': 'MNIST',
    'hb08ldm7': 'MNIST'
}

for run_id, dataset in runs_to_update.items():
    # Получить запуск
    run = wandb.Api().run(f"alexxandra_u/phtx/{run_id}")  # Замените username и project_name
    
    # Обновить конфигурацию
    run.config["dataset"] = dataset
    run.update()
```

```python
import pandas as pd


df_all = pd.read_csv('wandb_export_all.csv')
all_ids = list(df_all['ID'])
```

```python
params = {'d_model' : 64,
          'd_hidden' : 192,
          'num_heads': 8,
          'num_layers': 4,
          'dropout': 0}

for run_id in all_ids:
    for k, v in params.items(): 
        run = wandb.Api().run(f"alexxandra_u/phtx/{run_id}")
        run.config[k] = v
        run.update()
```
